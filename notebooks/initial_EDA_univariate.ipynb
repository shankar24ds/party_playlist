```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing the cleaned dataset
edf = pd.read_csv('../data/processed/cleaned_data_exploded.csv')
idf = pd.read_csv('../data/processed/cleaned_data_imploded.csv')
```

```python
idf.columns
```

```python
columns_to_plot = ['popularity', 'duration_s', 'explicit', 'danceability', 'energy',
                   'key', 'loudness', 'mode', 'speechiness', 'acousticness',
                   'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']

# Iterate through the columns and show each plot one at a time
for col in columns_to_plot:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    
    # Create histogram
    sns.histplot(data=idf, x=col, ax=axes[0], kde=True)
    axes[0].set_title(f'Histogram of {col}')
    
    # Create boxplot
    sns.boxplot(data=idf, y=col, ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()
```
